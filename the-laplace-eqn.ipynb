```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#%matplotlib notebook
```

```python
xvalues = np.linspace(0, 20, 20) 
yvalues = xvalues
```

```python
X, Y = np.meshgrid(xvalues, yvalues)
```

```python
Phi = np.zeros_like(X)
```

```python
# Boundary conditions

Phi[0, :] = 20                 #top boundary
Phi[-1, :] = 40                #bottom boundary
Phi[:, 0] = 10                 #bottom boundary
Phi[:, -1] = 30                #bottom boundary
```

```python
for k in range(500):
    for i in range(1, len(xvalues)-1):
        for j in range(1, len(yvalues)-1):
            Phi[i][j]  =  0.25 * (Phi[i+1][j] + Phi[i-1][j] + Phi[i][j+1] + Phi[i][j-1])
```

```python
colormap = plt.cm.jet
colorinterpolation = 50
```

```python
plt.contourf(X, Y, Phi, colorinterpolation, cmap = colormap)
cbar = plt.colorbar()
cbar.set_label('Volts ')
```

```python
ticklabels = [i for i in np.arange(0, 2.1, 0.25)]
plt.gca().xaxis.set_ticklabels(ticklabels)
plt.gca().yaxis.set_ticklabels(ticklabels)
plt.xlabel('Width of the domain (metres)')    
plt.ylabel('Height of the domain (metres)')    
```
